## Exercise 1: Prime Numbers


```python
def is_prime(n):
    if n <= 1:
        return False
    if n <= 3:
        return True
    # Check if divisible by 2 or 3
    if n % 2 == 0 or n % 3 == 0:
        return False
    # Check divisibility only for numbers of form 6kÂ±1 (optimization)
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

number = 17
result = is_prime(number)
print(f"{number} is prime: {result}")

number = 20
result = is_prime(number)
print(f"{number} is prime: {result}")
```

## Exercise 2: Product of Random Numbers


```python
import random

num1 = random.randint(1, 20)
num2 = random.randint(1, 20)
product = num1 * num2

print(f"First number: {num1}")
print(f"Second number: {num2}")
print(f"What is the product of {num1} and {num2}?")

user_answer = product

print(f"User entered: {user_answer}")
if user_answer == product:
    print("Correct! Well done.")
else:
    print(f"Incorrect. The correct answer is {product}.")
```

## Exercise 3: Squares of Even/Odd Numbers


```python
print("Squares of even numbers from 100 to 200:")

# Loop through numbers and check if even, then print square
for num in range(100, 201):
    if num % 2 == 0:
        print(f"{num}^2 = {num * num}")
```

## Exercise 4: Word Counter


```python
input_text = "This is a sample text. This text will be used to demonstrate the word counter."

words = input_text.split()
word_count = {}

# Count occurrences of each word using dictionary
for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

for word, count in word_count.items():
    print(f"'{word}': {count}")
```

## Exercise 5: Check for Palindrome


```python
def is_palindrome(s):
    # Remove spaces/punctuation and convert to lowercase, then compare with reverse
    cleaned = ''.join(char.lower() for char in s if char.isalnum())
    return cleaned == cleaned[::-1]

test_cases = ["racecar", "hello", "A man a plan a canal Panama", "12321", "Python"]

for test in test_cases:
    result = is_palindrome(test)
    print(f"Input: '{test}' -> Output: {result}")
```
